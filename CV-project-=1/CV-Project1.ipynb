```python
import cv2
import numpy as np
pic1=cv2.imread("-2.png")
resize1=cv2.resize(pic1,(800,800))
pic2=cv2.imread("-3.png")
resize2=cv2.resize(pic2,(800,800))
cv2.imshow("Pic1=",pic1)
cv2.waitKey(1000)
cv2.imshow("Pic2=",pic2)
cv2.waitKey(1000)
FinalImage=cv2.addWeighted(resize1,0.5,resize2,0.8,0)
cv2.imshow("FinalImage",FinalImage)
cv2.waitKey(5000)
cv2.destroyAllWindows()
print(cv2.__version__)
```
