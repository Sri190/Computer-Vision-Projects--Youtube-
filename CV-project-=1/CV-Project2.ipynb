```python
import cv2
import numpy as np
import face_recognition as fr

# Load known images
try:
    image1 = fr.load_image_file('image-1.jpg')
    image2 = fr.load_image_file('image-2.jpg')
except FileNotFoundError:
    print("âŒ One or both image files not found. Check filenames.")
    exit()

# Encode faces safely
encoding1_list = fr.face_encodings(image1)
encoding2_list = fr.face_encodings(image2)

if not encoding1_list or not encoding2_list:
    print("âŒ No face found in one or both images.")
    exit()

# First encoding from each image
encoding1 = encoding1_list[0]
encoding2 = encoding2_list[0]

known_encodings = [encoding1, encoding2]
known_names = ['Sri Hari Vathsan G', 'Sri Hari Vathsan G']

# Initialize webcam
video_cam = cv2.VideoCapture(0)

if not video_cam.isOpened():
    print("âŒ Cannot access webcam.")
    exit()

while True:
    ret, frame = video_cam.read()
    if not ret:
        print("âŒ Failed to read frame from webcam.")
        break

    # Convert frame from BGR to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect faces
    face_locations = fr.face_locations(rgb_frame)
    print("ğŸ§  Detected", len(face_locations), "face(s) in this frame.")

    # If no face detected, skip
    if not face_locations:
        cv2.imshow('Webcam Face Recognition', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        continue

    # Get encodings
    try:
        face_encodings = fr.face_encodings(rgb_frame, face_locations)
    except Exception as e:
        print("âŒ Error in encoding:", e)
        continue

    # Compare detected faces
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        matches = fr.compare_faces(known_encodings, face_encoding)
        distances = fr.face_distance(known_encodings, face_encoding)

        name = "Unknown"
        if len(distances) > 0:
            best_match_index = np.argmin(distances)
            if matches[best_match_index]:
                name = known_names[best_match_index]

        # Draw bounding box
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.rectangle(frame, (left, bottom q- 35), (right, bottom), (0, 255, 0), cv2.FILLED)
        cv2.putText(frame, name, (left + 6, bottom - 6),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 1)

    # Show output
    cv2.imshow('Webcam Face Recognition', frame)

    # Exit on 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
video_cam.release()
cv2.destroyAllWindows()
```
